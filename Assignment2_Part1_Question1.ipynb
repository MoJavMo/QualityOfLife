```python
# importing pandas, and other necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
happiness2017= pd.read_csv('~/Downloads/COMP309A2-Data/Part1/2017.csv', index_col=0)
happiness2018= pd.read_csv('~/Downloads/COMP309A2-Data/Part1/2018.csv', index_col=0)
happiness2019= pd.read_csv('~/Downloads/COMP309A2-Data/Part1/2019.csv', index_col=0)
```

```python
fig = plt.figure(figsize = (15,9))
sns.heatmap(happiness2017.corr(), cmap='Blues', annot = True);
```

```python
fig = plt.figure(figsize = (15,9))
sns.heatmap(happiness2018.corr(), cmap='Blues', annot = True);
```

```python
fig = plt.figure(figsize = (15,9))
sns.heatmap(happiness2019.corr(), cmap='Blues', annot = True);
```

```python
 fig, ax = plt.subplots(10, 2, figsize=(20,100))
for key, value in happiness2017.iteritems():
    index = happiness2017.columns.get_loc(key)
    if index == 0 or index == 1:
        continue
    ax=plt.subplot(10,2,2*(index-2)+1)
    ax.set_title(key+"(Top 20)")
    happiness2017[key].sort_values().tail(20).plot(kind="barh")
    ax=plt.subplot(10,2,2*(index-1))
    ax.set_title(key+"(Bottom 20)")
    happiness2017[key].sort_values().head(20).plot(kind="barh")
```

```python
 fig, ax = plt.subplots(7, 2, figsize=(20,100))
for key, value in happiness2018.iteritems():
    index = happiness2018.columns.get_loc(key)
    if index == 0 or index == 1:
        continue
    ax=plt.subplot(7,2,2*(index-2)+1)
    ax.set_title(key+"(Top 20)")
    happiness2018[key].sort_values().tail(20).plot(kind="barh")
    ax=plt.subplot(7,2,2*(index-1))
    ax.set_title(key+"(Bottom 20)")
    happiness2018[key].sort_values().head(20).plot(kind="barh")
```
